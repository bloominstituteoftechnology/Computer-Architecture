```python
def predictTemperature(startDate, endDate, temperature, n):
    days = []
    for hour in range(0, len(temperature), 24):
        days.append(temperature[hour:hour+24])
    average_day = np.mean(days, axis=0)
    predictions = []
    for _ in range(n):
        predictions.extend(average_day)
    return predictions
```
